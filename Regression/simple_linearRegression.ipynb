```python
import numpy as np

x = np.array([1,3,4,2,5], dtype=float)
y = np.array([3,4,5,2,1], dtype=float)

n = len(x)
x_mean = x.mean()
y_mean = y.mean()

b1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
b0 = y_mean - b1 * x_mean

def predict(x_new):
    return b0 + b1 * x_new

print("b0 =", b0, "b1 =", b1)
print("pred at x=6:", predict(6.0))
```
