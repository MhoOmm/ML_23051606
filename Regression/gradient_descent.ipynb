```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([6, 3, 4, 4, 6], dtype=float)

alpha = 0.01
iterations = 5

b0 = 1.0
b1 = 1.0

errors = []
b0_list = []
b1_list = []

for it in range(iterations):
    y_pred = b0 + b1 * x

    error = np.sum((y - y_pred) ** 2)
    errors.append(error)
    b0_list.append(b0)
    b1_list.append(b1)

    db0 = -2 * np.sum(y - y_pred)
    db1 = -2 * np.sum((y - y_pred) * x)

    b0 = b0 - alpha * db0
    b1 = b1 - alpha * db1

for i in range(iterations):
    print(f"Iter {i+1}: b0 = {b0_list[i]:.6f}, b1 = {b1_list[i]:.6f}, error = {errors[i]:.6f}")

plt.figure()
plt.plot(range(1, iterations + 1), errors, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Sum of Squared Errors')
plt.title('Error Curve for Gradient Descent (4 iterations)')
plt.grid(True)
plt.show()
```
